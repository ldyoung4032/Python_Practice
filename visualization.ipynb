```python
x = np.arange(1,10,0.1)
y = x*0.2
y2 = np.sin(x)

plt.plot(x,y,'b',label='first')
plt.plot(x,y2,'r',label='second')
plt.xlabel('x axis', size = 15)
plt.ylabel('y axis', size = 15)
plt.title('matplotlib sample')
plt.legend(loc='upper right')

plt.annotate('접점!', xy = (2.25, 0.5), xytext = (5,1),arrowprops=dict(facecolor='black', shrink=0.1), size = 50)
plt.show()
```

```python
t = np.arange(0,5,0.01)
plt.figure(figsize = (10,18))

plt.subplot(411)
plt.plot(t,np.sqrt(t))
plt.show()

plt.subplot(446)
plt.plot(t,t**2)
plt.show()
```

```python
t = np.arange(0,10,0.5)

plt.figure(figsize=(10,10))

plt.subplot(2,2,3)
plt.plot(t,t**2, c='g', linewidth = 10)
plt.grid()
```

```python
y = [2,3,1]
x = np.arange(len(y))
xlabel = ['가', '나', '다']
plt.title("막대그래프")
plt.bar(x,y)
plt.xticks(x,xlabel)
plt.yticks(sorted(y))
plt.xlabel("가나다")
plt.ylabel("빈도 수")
plt.show()
## 데이터.plot(kind = 'bar')
```

```python
np.random.seed(0)
people = ['a', 'v', 'r', 't']
y_pos = np.arange(len(people))
a = np.random.rand(len(people))
p = 1 + 10*a
plt.title("막대그래프")
plt.barh(y_pos, p, alpha = 0.4) ## horizontal
plt.yticks(y_pos, people)
plt.show()
```

```python
x=np.arange(15).reshape((5, 3))
print(x)
plt.barh(x[:,0], x[:,1], color='g', height=1.5) # 여기서 height는 bar 굵기!
plt.barh(x[:,0], -x[:,2], color='r', height=2)
plt.show()

## x[:,0] : 모든행 0열
```
