```python
## sns.relplot(x축 데이터,y축 데이터,data)
## sns.scatterplot(data,x축 데이터,y축 데이터)
sns.relplot(x = 'sepal_length', y = 'petal_length', data = iris, color = 'black')
```

```python
sns.scatterplot(x="sepal_length", y="petal_length", data=iris, color="green")
```

```python
## 연속형 변수와 범주형 변수가 함께 있는 경우 hue는 색상, style은 모양
sns.scatterplot(x = "sepal_length", y = "petal_length", data = iris, hue = "species")
```

```python
sns.scatterplot(x="sepal_length", y="petal_length",data=iris, style="species")
```

```python
## 기존 그래프에 실수형(연속형)데이터도 표시하고 싶을때
## hue, style 사용
sns.scatterplot(x="sepal_length", y="petal_length",data=iris,hue="sepal_width")
```

```python
sns.scatterplot(x="sepal_length", y="petal_length",data=iris,size="sepal_width", color="orange")
```

```python
##################그래프 두개
##범주에 따른 연속
sns.catplot(x = "species", y = "petal_length", data = iris)
## x종류에 따른 y
```

```python
sns.catplot(x = "species", y = "petal_length", data = iris, jitter = False)
# jitter 인자를 False로 설정하여 결과를 일직선으로 나타낼 수 있음
```

```python
## stripplot = scatterplot과 비슷
sns.stripplot(x = "species", y = "petal_length", data = iris)
```

```python
sns.stripplot(x="pclass", y="fare",data=titanic, hue="sex", jitter=True) 
# stripplot도 hue 추가 가능
```

```python
sns.stripplot(x="pclass", y="fare",data=titanic, hue="survived",
              jitter=True, dodge=True) 
## dod는 hue의 범주형 데이터들을 나눠서 시각화
## default는 False
```

```python
## swamplot = stripplot과 비슷하지만 데이터를 나타내는 점이 겹치지 않도록 옆으로 이동
sns.swarmplot(x = 'species', y = 'sepal_length', data = iris)
```

```python
## catplot = 범주(x)에 따른 연속(y)
sns.catplot(x = 'pclass', y = 'fare', hue = 'sex', data = titanic)
```

```python
### boxplot = catplot인데 kind = 'box'해줘야한다
##박스는 실수 값 분포에서 1사분위수(Q1)와 3사분위수(Q3)를 뜻하고, 3사분위수와 1사분수의 차이(Q3 - Q1)는 IQR이라고 한다.
##1사분위 수보다 1.5 x IQR 만큼 낮은 값과 3사분위 수보다 1.5 x IQR 만큼 높은 값의 구간을 기준으로 그 바깥의 점은 outlier(이상치)이다
```

```python
sns.catplot(x = 'species', y = 'petal_length', kind = 'box', data = iris)
```

```python
sns.catplot(x="pclass", y="fare", hue="sex", kind="box", data=titanic) #hue 파라미터 추가
```

```python
## violin plot = boxplot보다 자세하게
##boxplot은 분포의 간략한 특성만 보여주지만 violinplot은 카테고리값에 따른 각 분포의 실제 데이터나 전체 형상을 보여준다는 장점이 있다.
##violinplot은 세로 방향으로 커널 밀도 히스토그램을 그려주는데 왼쪽과 오른쪽이 대칭이 되도록 하여 바이올린처럼 보인다고 하여 붙은 이름이다
```

```python
sns.violinplot(x = 'month', y = 'passengers', 
               data = flights)

## 중앙값, 1분위수, 3분위수, min, max
## 커널밀도추정값을 붙일 수 있다

sns.set_palette("pastel")
```

```python
sns.violinplot(x="pclass", y="fare", hue="sex", data=titanic) # hue 파라미터 추가 가능
```

```python
sns.violinplot(x="pclass", y="fare", hue="sex", data=titanic, split=True)
## hue가 두가지 범주형 데이터이면 split=True를 사용하여 반반으로
## 그래프화 가능
```

```python
## barplot
## 카테고리 값에 따른 실수 값의 (평균과 편차)를 표시하는 기본적인 바 차트를 생성
sns.barplot(x = 'pclass', y = 'fare', data = titanic)
```

```python
sns.catplot(x="sex", y="survived", hue="class", kind="bar", data=titanic) #catplot에 kind="bar"를 이용
```

```python
## rug plot
## 데이터 위치를 x축 위에 작은 선분(rug)로 나타내어 실제 데이터들의 위치를 보여줌
sns.rugplot(iris.sepal_length.values)

## rugplot의 y축은 무엇을 나타내는가?
```

```python
iris.sepal_length.values
```

```python
## KDe Plot = 히스토그램보다 부드러운 곡선
sns.kdeplot(iris.sepal_length.values, color = 'r')
```

```python
plt.hist(iris.sepal_length.values, color = 'r')
```

```python
## disfplot
## rug + 커널밀도 + 히스토그램 표시
sns.distplot(iris.sepal_length.values, kde=True, rug = True, color = "r")
```

```python
##count plot
## 각 카테고리 값 별로 테이터가 얼마나 있는지 표시 가능
## 빈도수를 그래프에 표시
sns.countplot(x = "pclass", data = titanic)
```

```python
sns.countplot(x = 'survived', data = titanic)
```

```python
## jointplot
## 스캐터 플롯 + 차트의 가장 자리에 각 변수의 히스토그램도 그린다.
sns.jointplot(x = 'petal_length', y ='petal_width', data = iris, color = 'g')
```
